```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import os

# Absolute path
df = pd.read_csv("/content/sample_data/multiple_linear_regression_dataset.csv")

```

```python
df.head()
```

```python
cdf = df[['age','experience','income']]
cdf.head(9)
```

```python
plt.scatter(cdf.age, cdf.income,  color='blue')
```

```python
plt.scatter(cdf.experience, cdf.income,  color='red')
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['age','experience']])
y = np.asanyarray(train[['income']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
y_hat= regr.predict(test[['age','experience']])
x = np.asanyarray(test[['age','experience']])
y = np.asanyarray(test[['income']])
print("Mean Squared Error (MSE) : %.2f"
      % np.mean((y_hat - y) ** 2))

print('Variance score: %.2f' % regr.score(x, y))
```
